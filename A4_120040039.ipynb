```python
import numpy as np
from random import *

data = np.loadtxt("seeds_dataset.txt")
new_data = data[:,:-1]
rslt = []
for i in range(len(data)):
    rslt.append(data[i][7])
    
    
def E_D(x, y):
    x = np.array(x)
    y = np.array(y)
    return np.sqrt(np.sum((x-y)**2))


data = new_data
rslt
```

```python
SEs = []
RIs = []
its = []

for p in range (10):
    m, n = np.shape(data)
    centroids = np.zeros((3,n))
    for i in range(3):
        index = int(np.random.uniform(0,m)) 
        centroids[i,:] = data[index,:]
    c_ass = np.mat(np.zeros((m, 2)))
    check = True
    iteration = 0
    while check:
        iteration += 1
        check = False
        for i in range(m):
            mindis = 1e5
            minindex = -1
            for j in range(3):
                x1 = np.array(centroids[j,:])
                x2 = np.array(data[i,:])
                distJ = np.sqrt(np.sum((x1 - x2) ** 2))
                if distJ < mindis:
                    mindis = distJ
                    minindex = j
            if c_ass[i, 0] != minindex:
                check = True
            c_ass[i,:] = int(minindex), mindis
        for cent in range(3):
            ptsinclust = data[np.nonzero(c_ass[:,0].A == cent)[0]]
            centroids[cent,:] = np.mean(ptsinclust, axis= 0)
    centroids, c_ass, iteration

    SE=[]
    for t in range(10):
        avg1 = []
        avg2 = []
        avg3 = []


        point = randint(0,m-1)
        for i in range(3):
            if i != c_ass[point, 0]:
                clst2 = i
                break
        for i in range(3):
            if i != c_ass[point, 0] and i != clst2:
                clst3 = i
                break
            

        for i in range(m):
            if c_ass[i, 0] == c_ass[point, 0]:
                dis=E_D(data[i,:],data[point,:])
                avg1.append(dis)
            elif c_ass[i, 0] == clst2:
                dis=E_D(data[i,:],data[point,:])
                avg2.append(dis)
            else:
                dis=E_D(data[i,:],data[point,:])
                avg3.append(dis)

        ai=np.mean(avg1)
        a = np.mean(avg2)
        b = np.mean(avg3)
        if a > b:
            bi = b
        else:
            bi = a
        SE.append(((bi-ai)/max(ai,bi)))
    print(SE)
    avgSE=np.mean(SE)
    avgSE

    lA = 0
    lB = 0
    lC = 0
    lD = 0
    for i in range(m):
        for j in range(m):
            if rslt[i] == rslt[j] and c_ass[i, 0] == c_ass[j, 0]:
                lA += 1
            elif rslt[i] != rslt[j] and c_ass[i, 0] != c_ass[j, 0]:
                lB += 1
            elif rslt[i] == rslt[j] and c_ass[i, 0] != c_ass[j, 0]:
                lC += 1
            elif rslt[i] != rslt[j] and c_ass[i, 0] == c_ass[j, 0]:
                lD += 1

    RI = (lA+lB) / (lA+lB+lC+lD)
    RI
    
    SEs.append(avgSE)
    RIs.append(RI)
    its.append(iteration)
    
    
print('SE: ',SEs)

print('RI: ',RIs)
print('Iteration ', its)
print('Variance of SE \t',np.var(SEs))
print('Mean of SE \t',np.mean(SEs))
print('Variance of RI \t',np.var(RIs))
print('Mean of RI :\t',np.mean(RIs))
```
